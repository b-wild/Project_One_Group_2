### Data Exploration and Cleanup - Merging Weather and Crime Data

```python
# access crime data and weather data
weather_file_path = "output_data/dc_weather_by_hour.csv"
crime_file_path = "Resources/clean_crime_data.csv"

# create a Pandas DataFrame for each file
weather_df = pd.read_csv(weather_file_path)
crime_df = pd.read_csv(crime_file_path)
```

```python
# Review sample of weather data
weather_df.head()
```

```python
# Review sample of crime data
crime_df.head()
```

```python
# Convert crime date type to match weather date type
crime_df['START_DATE'] = pd.to_datetime(crime_df['START_DATE'], format='%Y/%m/%d').dt.strftime('%Y-%m-%d')
# Check sample to see results
crime_df.head()
```

```python
# Parse out hour from START_TIME
crime_df[['START_HOUR','START_MIN_SEC']] = crime_df['START_TIME'].str.split(':',n=1,expand=True)
crime_df.head()
```

```python
# Create an hour list based on 'START_HOUR' to match 00:00AM/PM weather hour format
# Make a list out of 'START_HOUR' column.
hour_num_list = crime_df["START_HOUR"].values.tolist()

# Create a new list for the hours
hour_list = []

# Iterate over the 'START_HOUR' column to create hour list.
for hour in hour_num_list:
    if hour == "00":
        hour_list.append("12:00AM")
    elif hour == "01":
        hour_list.append("1:00AM")
    elif hour == "02":
        hour_list.append("2:00AM")
    elif hour == "03":
        hour_list.append("3:00AM")
    elif hour == "04":
        hour_list.append("4:00AM")
    elif hour == "05":
        hour_list.append("5:00AM")
    elif hour == "06":
        hour_list.append("6:00AM")
    elif hour == "07":
        hour_list.append("7:00AM")
    elif hour == "08":
        hour_list.append("8:00AM")
    elif hour == "09":
        hour_list.append("9:00AM")
    elif hour == "10":
        hour_list.append("10:00AM")
    elif hour == "11":
        hour_list.append("11:00AM")
    elif hour == "12":
        hour_list.append("12:00PM")
    elif hour == "13":
        hour_list.append("1:00PM")
    elif hour == "14":
        hour_list.append("2:00PM")
    elif hour == "15":
        hour_list.append("3:00PM")
    elif hour == "16":
        hour_list.append("4:00PM")
    elif hour == "17":
        hour_list.append("5:00PM")
    elif hour == "18":
        hour_list.append("6:00PM")
    elif hour == "19":
        hour_list.append("7:00PM")
    elif hour == "20":
        hour_list.append("8:00PM")
    elif hour == "21":
        hour_list.append("9:00PM")
    elif hour == "22":
        hour_list.append("10:00PM")
    elif hour == "23":
        hour_list.append("11:00PM")

# Add new hour list to DataFrame
crime_df['Hour'] = hour_list
crime_df.head()
```

```python
crime_df = crime_df[['YEAR','START_DATE','START_TIME','Hour','offensegroup','OFFENSE','offense-text']]
crime_df.head()
```

```python
# Merge crime dataframe with weather dataframe
crime_weather = pd.merge(crime_df, weather_df, how='left', left_on=['START_DATE','Hour'], right_on=['date','hour'])
# Remove all rows outside of date range
crime_weather.dropna(how='any', inplace=True)
# Sort by start date and time
crime_weather.sort_values(["date", "START_TIME"], ascending=True, inplace=True)
# Reset Index
crime_weather.reset_index(drop=True, inplace=True)
# Remove redundant and unnecessary columns
crime_weather = crime_weather[['YEAR', 'START_DATE', 'START_TIME', 'Hour', 'offensegroup', 'OFFENSE',
       'offense-text','maxtemp_f', 'mintemp_f','avgtemp_f', 'totalprecip_in', 'avghumidity', 'day_condition',
       'hour_temp_f', 'hour_condition', 'hour_precip_in', 'hour_humidity']]
# Rename columns for cleaner presentation
crime_weather.rename(columns={"YEAR": "Year", "START_DATE": "Start Date", "START_TIME": "Start Time",
                    "offensegroup": "Offense Group", 'OFFENSE': 'Offense', 'offense-text': 'Offense Text',
                    'maxtemp_f': 'Max Temp (f)', 'mintemp_f': 'Min Temp (f)', 'avgtemp_f': 'Avg Temp (f)',
                    'totalprecip_in': 'Total Precipitation (in)', 'avghumidity': 'Avg Humidity (%)',
                    'day_condition': 'Day Condition', 'hour_temp_f': 'Hour Temp (f)', 'hour_condition': 'Hour Condition',
                    'hour_precip_in': 'Hour Precipitation (in)','hour_humidity': 'Hour Humidity (%)'}, inplace=True)
# View results
crime_weather
```

```python
# Export the merged dataframe
crime_weather.to_csv("output_data/crime_weather_merged.csv", index_label="ID")
```
